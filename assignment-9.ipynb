Q6. List down some best practices for exception handling in python.

#Ans.
1.Use specific exception types: Catch specific exception types instead of using a generic 'except' clause. This allows you to handle different exceptions differently and provides better clarity.

2.Use 'try-except' blocks: Wrap the code that can potentially raise an exception within a try-except block. This allows you to catch and handle exceptions gracefully.

3.Handle exceptions gracefully: When catching exceptions, handle them in a way that provides useful information to the user or the developer. Display helpful error messages or take appropriate actions to recover from the exception.

4.Avoid bare 'except' clauses: Avoid using bare except clauses (without specifying the exception type) as they catch all exceptions, including system-exiting exceptions. This can hide critical errors and make debugging difficult. Instead, catch specific exception types or use a generic except Exception clause.
